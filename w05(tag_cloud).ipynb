```
def fib1(n):
  if n==1 or n==2:
    return 1
  return fib1(n-1)+fib1(n-2)


%timeit fib1(33)
%timeit fib1(34)
%timeit fib1(37)
```

```
#custom factorail
import math
def factorial(n):
  ret = 1
  for i in range(1,n+1):
    ret = ret * i
  return ret

print(factorial(11)==math.factorial(11))
print(factorial(12)==math.factorial(12))
print(factorial(13)==math.factorial(13))
```

```
def fib2(n): 
    a=0
    b=1
    for i in range(n-1):
        b, a = a+b, b
    return b
      
%timeit fib2(33)
%timeit fib2(34)
%timeit fib2(370)
```
